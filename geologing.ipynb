```python
from geopy.geocoders import ArcGIS
import pandas as pd
```

```python
nom = ArcGIS()
```

```python
df_kecamatan = pd.read_csv("LatLong.csv")
```

```python
df_kecamatan
```

```python
data = []
```

```python
for index, row in df_kecamatan.iterrows():
    lokasi = nom.geocode(row["Nama Kecamatan"])  # Sesuaikan dengan nama kolom di CSV Anda
    if lokasi:
        data.append({
            "Kecamatan": row["Nama Kecamatan"],  # Sesuaikan dengan nama kolom di CSV Anda
            "Latitude": lokasi.latitude,
            "Longitude": lokasi.longitude
        })
    else:
        print(f"Tidak dapat menemukan koordinat untuk kecamatan: {row['Nama_Kecamatan']}")
```

```python
df_hasil = pd.DataFrame(data)
```

```python
print(df_hasil)
```
